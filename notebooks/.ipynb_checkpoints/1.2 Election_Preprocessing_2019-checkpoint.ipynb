```python
# Data taken from https://www.elections.ca/content.aspx?section=res&dir=rep/off/42gedata&document=byed&lang=e
```

```python
import pandas as pd
from glob import glob
import os
import numpy as np
```

```python
directory=os.getcwd().replace('notebooks','data\\raw\\elections\\2015\\')
provinces=next(os.walk(directory))[1]
df_list=[]
for province in provinces:
    province_directory=f'{directory}{province}'
    files=glob(f'{province_directory}\*.csv')
    for file in files:
        df = pd.read_csv(file,index_col=None,header=0)
        df['Province']=province
        df_list.append(df)
election_df=pd.concat(df_list,axis=0,ignore_index=True)
election_df.head(100)
```

```python
column_names=[]
for column in election_df.columns:
    column_name=column.split('/')[0].replace(' ','_')
    column_names.append(column_name)
election_df.columns=column_names
election_df.head()
```

```python
election_df.Void_Poll_Indicator.value_counts()
```

```python
election_df.Political_Affiliation_Name_English.value_counts()
```

```python
election_df['Political_Party']=np.select(
    [
        election_df.Political_Affiliation_Name_English == 'Liberal', 
        election_df.Political_Affiliation_Name_English == 'Conservative',
        election_df.Political_Affiliation_Name_English == 'NDP-New Democratic Party',
        election_df.Political_Affiliation_Name_English == 'Green Party',
        election_df.Political_Affiliation_Name_English == 'Bloc Québécois'
    ], 
                        
    [
        'Liberal', 
        'Conservative',
        'NDP-New Democratic Party',
        'Green Party',
        'Bloc Québécois'
    ], 
default='Other')
election_df['Political_Party']=election_df.Political_Party.str.replace(' ','-')
```

```python
election_df.head()
```

```python
polling_df=pd.DataFrame(election_df.groupby(['Electoral_District_Number','Province','Polling_Station_Number']).agg({'Electors_for_Polling_Station':'max'})).reset_index()
polling_df
```

```python
district_df=pd.DataFrame(polling_df.groupby(['Electoral_District_Number','Province']).agg(Number_of_Polling_Stations=('Polling_Station_Number','nunique'),Number_of_Electors=('Electors_for_Polling_Station','sum'))).reset_index()
district_df
```

```python
votes_df=pd.DataFrame(election_df.groupby(['Electoral_District_Number','Province','Political_Party'])['Candidate_Poll_Votes_Count'].sum()).reset_index()
votes_df
```

```python
votes_df['Vote_Rank']='Position_'+votes_df.groupby(['Electoral_District_Number'],sort=False)["Candidate_Poll_Votes_Count"].rank(method='dense',ascending=False,axis=0).astype(int).astype(str)
votes_df
```

```python
ranking_df=pd.pivot(votes_df,columns='Vote_Rank',index=['Electoral_District_Number','Province'],values='Political_Party').fillna("None").reset_index()
ranking_df
```

```python
political_ranking_df=pd.pivot(votes_df,columns='Political_Party',index=['Electoral_District_Number','Province'],values='Vote_Rank').reset_index()
political_ranking_df['Bloc-Québécois']=political_ranking_df['Bloc-Québécois'].str.replace('Position_','').astype(float)
political_ranking_df['Conservative']=political_ranking_df['Conservative'].str.replace('Position_','').astype(float)
political_ranking_df['Green-Party']=political_ranking_df['Green-Party'].str.replace('Position_','').astype(float)
political_ranking_df['Liberal']=political_ranking_df['Liberal'].str.replace('Position_','').astype(float)
political_ranking_df['NDP-New-Democratic-Party']=political_ranking_df['NDP-New-Democratic-Party'].str.replace('Position_','').astype(float)
political_ranking_df['Other']=political_ranking_df['Other'].str.replace('Position_','').astype(float)
political_ranking_df['Temp_Max_Rank']=political_ranking_df[['Bloc-Québécois','Conservative','Green-Party','Liberal','NDP-New-Democratic-Party','Other']].max(axis=1)

political_ranking_df.loc[political_ranking_df['Bloc-Québécois'].isnull(),'Bloc-Québécois'] = political_ranking_df['Temp_Max_Rank']+1
political_ranking_df.loc[political_ranking_df['Conservative'].isnull(),'Conservative'] = political_ranking_df['Temp_Max_Rank']+1
political_ranking_df.loc[political_ranking_df['Green-Party'].isnull(),'Green-Party'] = political_ranking_df['Temp_Max_Rank']+1
political_ranking_df.loc[political_ranking_df['Liberal'].isnull(),'Liberal'] = political_ranking_df['Temp_Max_Rank']+1
political_ranking_df.loc[political_ranking_df['Other'].isnull(),'Other'] = political_ranking_df['Temp_Max_Rank']+1
political_ranking_df.loc[political_ranking_df['NDP-New-Democratic-Party'].isnull(),'NDP-New-Democratic-Party'] = political_ranking_df['Temp_Max_Rank']+1
del political_ranking_df['Temp_Max_Rank']
political_ranking_df=political_ranking_df.set_index(['Electoral_District_Number','Province'])
political_ranking_df=political_ranking_df.astype(int)
political_ranking_df=political_ranking_df.reset_index()
political_ranking_df
```

```python
ranking_df['Position_1'].value_counts()
```

```python
party_votes_df=pd.pivot(votes_df,values='Candidate_Poll_Votes_Count',index=['Electoral_District_Number','Province'],columns='Political_Party').reset_index()
party_votes_df=party_votes_df.fillna(0)
party_votes_df=party_votes_df.astype({'Bloc-Québécois': int, 'Conservative': int,'Green-Party': int, 'Liberal': int,'NDP-New-Democratic-Party': int, 'Other': int})
party_votes_df['Vote_Total']=party_votes_df['Bloc-Québécois']+party_votes_df['Conservative']+party_votes_df['Green-Party']+party_votes_df['Liberal']+party_votes_df['NDP-New-Democratic-Party']+party_votes_df['Other']
party_votes_df.rename(columns={'Bloc-Québécois': 'Bloc-Québécois_Votes', 'Conservative': 'Conservative_Votes','Green-Party': 'Green-Party_Votes', 'Liberal': 'Liberal_Votes','NDP-New-Democratic-Party': 'NDP-New-Democratic-Party_Votes', 'Other': 'Other_Votes'}, inplace=True)
party_votes_df
```

```python
pd.options.display.max_columns = None
final_df=(party_votes_df
         .merge(ranking_df,how='inner',on=['Electoral_District_Number','Province'])
         .merge(district_df,how='inner',on=['Electoral_District_Number','Province'])
         .merge(political_ranking_df,how='inner',on=['Electoral_District_Number','Province']))
final_df['Percent_Riding_Votes_Liberal']=final_df['Liberal_Votes'].fillna(0)/final_df['Vote_Total']
final_df['Percent_Riding_Votes_Bloc-Québécois']=final_df['Bloc-Québécois_Votes'].fillna(0)/final_df['Vote_Total']
final_df['Percent_Riding_Votes_Conservative']=final_df['Conservative_Votes'].fillna(0)/final_df['Vote_Total']
final_df['Percent_Riding_Votes_Green-Party']=final_df['Green-Party_Votes'].fillna(0)/final_df['Vote_Total']
final_df['Percent_Riding_Votes_NDP-New-Democratic-Party']=final_df['NDP-New-Democratic-Party_Votes'].fillna(0)/final_df['Vote_Total']
final_df['Percent_Riding_Votes_Other']=final_df['Other_Votes'].fillna(0)/final_df['Vote_Total']

final_df['Voter_Participation']=final_df['Vote_Total']/final_df['Number_of_Electors']
final_df=final_df.set_index(['Electoral_District_Number','Province'])
final_df=final_df.add_prefix('2015_')
final_df=final_df.reset_index()
final_df.head()
```
