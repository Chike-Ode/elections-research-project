```python
import pandas as pd
from glob import glob
import os
import numpy as np
```

```python
directory=os.getcwd().replace('notebooks','data\\raw\\elections\\2019\\')
provinces=next(os.walk(directory))[1]
df_list=[]
for province in provinces:
    province_directory=f'{directory}{province}'
    files=glob(f'{province_directory}\*.csv')
    for file in files:
        df = pd.read_csv(file,index_col=None,header=0)
        df['Province']=province
        df_list.append(df)
election_df=pd.concat(df_list,axis=0,ignore_index=True)
election_df.head(100)
```

```python
column_names=[]
for column in election_df.columns:
    column_name=column.split('/')[0].replace(' ','_')
    column_names.append(column_name)
election_df.columns=column_names
election_df.head()
```

```python
election_df.Void_Poll_Indicator.value_counts()
```

```python
election_df.Political_Affiliation_Name_English.value_counts()
```

```python
election_df['Political_Party']=np.select(
    [
        election_df.Political_Affiliation_Name_English == 'Liberal', 
        election_df.Political_Affiliation_Name_English == 'Conservative',
        election_df.Political_Affiliation_Name_English == 'NDP-New Democratic Party',
        election_df.Political_Affiliation_Name_English == 'Green Party',
        election_df.Political_Affiliation_Name_English == 'Bloc Québécois'
    ], 
                        
    [
        'Liberal', 
        'Conservative',
        'NDP-New Democratic Party',
        'Green Party',
        'Bloc Québécois'
    ], 
default='Other')
election_df['Political_Party']=election_df.Political_Party.str.replace(' ','-')
```

```python
election_df.head()
```

```python
polling_df=pd.DataFrame(election_df.groupby(['Electoral_District_Number','Province','Polling_Station_Number']).agg({'Electors_for_Polling_Station':'max'})).reset_index()
polling_df
```

```python
district_df=pd.DataFrame(polling_df.groupby(['Electoral_District_Number','Province']).agg(Number_of_Polling_Stations=('Polling_Station_Number','nunique'),Number_of_Electors=('Electors_for_Polling_Station','sum'))).reset_index()
district_df
```

```python
votes_df=pd.DataFrame(election_df.groupby(['Electoral_District_Number','Province','Political_Party'])['Candidate_Poll_Votes_Count'].sum()).reset_index()
votes_df
```

```python
votes_df['Vote_Rank']='Position_'+votes_df.groupby(['Electoral_District_Number'],sort=False)["Candidate_Poll_Votes_Count"].rank(method='dense',ascending=False,axis=0).astype(int).astype(str)
votes_df
```

```python
ranking_df=pd.pivot(votes_df,columns='Vote_Rank',index=['Electoral_District_Number','Province'],values='Political_Party').reset_index()
ranking_df
```

```python
ranking_df['Position_1'].value_counts()
```

```python
party_votes_df=pd.pivot(votes_df,values='Candidate_Poll_Votes_Count',index=['Electoral_District_Number','Province'],columns='Political_Party').reset_index()
party_votes_df=party_votes_df.fillna(0)
party_votes_df['Vote_Total']=party_votes_df['Bloc-Québécois']+party_votes_df['Conservative']+party_votes_df['Green-Party']+party_votes_df['Liberal']+party_votes_df['NDP-New-Democratic-Party']+party_votes_df['Other']
party_votes_df
```

```python
final_df=(party_votes_df
         .merge(ranking_df,how='inner',on=['Electoral_District_Number','Province'])
         .merge(district_df,how='inner',on=['Electoral_District_Number','Province']))
final_df['Percent_Liberal']=final_df['Liberal']/final_df['Vote_Total']
final_df['Percent_Bloc-Québécois']=final_df['Bloc-Québécois']/final_df['Vote_Total']
final_df['Percent_Conservative']=final_df['Conservative']/final_df['Vote_Total']
final_df['Percent_Green-Party']=final_df['Green-Party']/final_df['Vote_Total']
final_df['Percent_NDP-New-Democratic-Party']=final_df['NDP-New-Democratic-Party']/final_df['Vote_Total']
final_df['Percent_Other']=final_df['Other']/final_df['Vote_Total']

final_df['Voter_Participation']=final_df['Vote_Total']/final_df['Number_of_Electors']
final_df=final_df.set_index(['Electoral_District_Number','Province'])
final_df=final_df.add_prefix('2019_')
final_df=final_df.reset_index()
final_df.head()
```

```python
export_directory=directory.replace('\\raw\\elections\\2019\\','\\output\\elections\\')
final_df.to_csv(f'{export_directory}2019_election_results.csv',index=False)
```
