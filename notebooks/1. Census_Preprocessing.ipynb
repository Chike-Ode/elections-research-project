```python
def clean_column_names(df,clean_numbers=False,clean_percentages=True):
    if clean_percentages==True:
        df.columns = df.columns.str.replace('%', '')
    if clean_numbers==True:
        df.columns = df.columns.str.replace(r'\d+ ', '')
    df.columns = df.columns.str.replace(' ', '_')
    df.columns = df.columns.str.replace(',', '')
    df.columns = df.columns.str.lower()
    return df

def get_pivot_data(fields,total_field=None,data_point_dict=data_point_dict,total_df=total_df,return_both=True,clean_numbers=False,clean_percentages=True):
    df=pd.DataFrame()
    percent_df=pd.DataFrame(index=df.index.copy())
    if total_field is not None:
        total_field_name=data_point_dict.get(total_field)
    field_names=list(map(data_point_dict.get, fields))
    df[field_names]=total_df[fields].astype(float)
    columns=df.columns.values.tolist()
    if total_field is not None:
        columns.remove(total_field_name)
    for column in columns:
        if total_field is not None:
            percent_df[column]=df[column]/df[total_field_name]
    df=clean_column_names(df)
    percent_df=clean_column_names(percent_df,clean_numbers,clean_percentages)
    if return_both==True:
        return df.reset_index(),percent_df.reset_index()
    else:
        return df.reset_index()
    
def get_df_name(df):
    name =[x for x in globals() if globals()[x] is df][0]
    return name
def merge_dataframes(dataframe_list,prefix='percent_'):
    returned_df=pd.DataFrame()
    for index,dataframe in enumerate(dataframe_list):
        remove_prefix_str=get_df_name(dataframe)[len(prefix):]
        column_prefix=remove_prefix_str[:-len('df')]
        dataframe=dataframe.set_index('ALT_GEO_CODE')
        dataframe=dataframe.add_prefix(column_prefix)
        dataframe=dataframe.reset_index()
        if index==0:
            returned_df=dataframe
        else:
            returned_df=returned_df.merge(dataframe,how='inner',on='ALT_GEO_CODE')
    return returned_df
```
