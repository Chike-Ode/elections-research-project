```python
import pandas as pd
```

```python
import os
directory=os.getcwd().replace('notebooks','data\\raw\\')
filename="raw_census_data_2016.csv"
full_name=f"{directory}{filename}"
df=pd.read_csv(full_name)
df.head()
```

```python
districts_df=df[df['GEO_LEVEL']==2]
districts_df
```

```python
geo_mapping_df=districts_df[['ALT_GEO_CODE','GEO_NAME']].drop_duplicates()#.set_index('ALT_GEO_CODE').to_dict()
geo_mapping_dict=dict(zip(districts_df.ALT_GEO_CODE, districts_df.GEO_NAME))
```

```python
data_point_df=districts_df[['Member ID: Profile of Federal Electoral Districts (2013 Representation Order) (2247)','DIM: Profile of Federal Electoral Districts (2013 Representation Order) (2247)']].drop_duplicates()
data_point_dict=dict(zip(districts_df['Member ID: Profile of Federal Electoral Districts (2013 Representation Order) (2247)'], districts_df['DIM: Profile of Federal Electoral Districts (2013 Representation Order) (2247)']))
```

```python
# df.sales = df.sales.astype(str)
total_df=pd.pivot(districts_df,values='Dim: Sex (3): Member ID: [1]: Total - Sex',index='ALT_GEO_CODE',columns='Member ID: Profile of Federal Electoral Districts (2013 Representation Order) (2247)')#,aggfunc='sum')
total_df
```

```python
income_correlation_matrix=total_df[[751,752,753,754,755,756,757,758,759,800,801,802,804,805,806,808,809,810,812,813,814]].astype('float64').corr()
income_correlation_matrix
```

```python
income_correlation_matrix.mean().sort_values(ascending=False).rename(index=data_point_dict).to_frame()
```

```python
black_immigrant_birth_df=pd.DataFrame()
countries=[1162,1163,1164,1167,1188]
fields=[1324,1327,1338]
field_names=list(map(data_point_dict.get, fields))
black_immigrant_birth_df[field_names]=total_df[fields].astype(int)
black_immigrant_birth_df['black_immigrants']=total_df[countries].astype(int).sum(axis=1).to_frame()#,
print(f'Black immigrants are considered to be from the following countries: {" ".join(list(map(data_point_dict.get, countries)))}')
black_immigrant_birth_df.rename(index=geo_mapping_dict).sort_values('Black',ascending=False).head(50)
```
